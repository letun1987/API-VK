```python
import pandas as pd
import time
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
token = #здесь должен быть токен)))
version = 5.131
domain = 'yandextasks'
owner_id = -221651103 # номер владельца записи
post_id = 12 # номер нужной записи
```

```python
response = requests.get('https://api.vk.com/method/wall.get', # получаем доступ к записям на стене
                        params={
                            'access_token': token,
                            'v': version,
                            'domain': domain
                        })
data = response.json()
```

```python
response = requests.get('https://api.vk.com/method/wall.getComments', # получаем доступ к нужной записи и достаем комментарии
                        params={
                            'access_token': token,
                            'v': version,
                            'owner_id': owner_id,
                            'post_id': post_id,
                            'count': 100
                        })
data = response.json()['response']['items']
```

```python
comments = [] # создаем пустой список для комментариев
```

```python
for i in data:
  i = i.get('text')
  i = re.sub(r'[^А-яЁё]+', ' ', i).lower()
  i = ' '.join(i.split())
  if len(i)==0:
    pass
  else:
    comments.append(i) # добавляем комментарии в список, предварительно их почистив
```

```python
!pip install dostoevsky # для анализа тональности используем библиотеку Достоевский
!python3 -m dostoevsky download fasttext-social-network-model # загружаем размеченную модель
from dostoevsky.tokenization import RegexTokenizer
from dostoevsky.models import FastTextSocialNetworkModel
```

```python
tokenizer = RegexTokenizer()
model = FastTextSocialNetworkModel(tokenizer=tokenizer)

df = pd.DataFrame()

results = model.predict(comments, k=2)
for comments, sentiment in zip(comments, results):
    #print(comments, '-&gt;', sentiment)
    row = {'Комментарий': comments, 'Прогноз': sentiment}
    df = pd.concat([df, pd.DataFrame([row])]).reset_index(drop=True) #создаем датафрейм

df.head(20)
```
